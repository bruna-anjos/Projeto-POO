```python
package Funcionario;

import java.util.ArrayList;
import java.util.Scanner;

public class Dados {

	@SuppressWarnings("resource")
	public static void main(String[] args) {

		Scanner scanner = new Scanner(System.in);
		ArrayList<Funcionario> funcionarios = new ArrayList<Funcionario>();

		boolean rodando = true;
		while (rodando) {
			System.out.println("\n***Bem Vindo ao Cadastro Saúde Funcionarios ***\n");
			System.out.println("[1] - Cadastrar  Funcionário");
			System.out.println("[2] - Pesquisar Funcionário");
			System.out.println("[3] - Sair");
			System.out.println("\n Informe sua opção:");
			String opcao = scanner.nextLine();

			switch (opcao) {

				case "1": {
					System.out.println("===========CADASTRO DE FUNCIONÁRIOS ============");
					System.out.print("Codigo: ");
					String codigo = scanner.nextLine();
					System.out.print("Digite o nome: ");
					String nome = scanner.nextLine();
					System.out.print("Digite a idade: ");
					String idade = scanner.nextLine();
					System.out.print("Digite o telefone: ");
					String telefone = scanner.nextLine();
					System.out.print("Digite doença: ");
					String doenca = scanner.nextLine();
					Funcionario f = new Funcionario();
					f.setCodigo (codigo);
					f.setNome (nome);
					f.setIdade(idade);
					f.setTelefone(telefone);
					f.setdoenca(doenca);

					funcionarios.add(f);
					break;
				}
				case "2":{
					System.out.println("===========PROCURAR POR FUNCIONÁRIO ============");
					System.out.print("Digite o codigo: ");
					String codigo = scanner.nextLine();

					boolean encontrado = false;
					for (int i = 0; i < funcionarios.size(); i++) {

						Funcionario fTemp = funcionarios.get(i);

						System.out.println("Funcionário encontrado");
						System.out.println("\tCodigo..:" + fTemp.getCodigo());
						System.out.println("\tNome....:" + fTemp.getNome());
						System.out.println("\tIdade...:" + fTemp.getIdade());
						System.out.println("\tTelefone:" + fTemp.getTelefone());
						System.out.println("\tDoença..:" + fTemp.getDoenca());
						encontrado = true;
						break;
					}
	       			if (encontrado == false) {
	       				System.out.println("Não foi encotrado nenhum funcionário com esse código.");
	       			}
	       			break;
			    }
	       		case "3": {
	       			rodando = false;
	       			System.out.println("Cadastro encerrado.");
	       			break;
	       		}
			}
		}
	}
}
```
